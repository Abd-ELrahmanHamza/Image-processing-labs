# Reports
- Roberts:
    noisy image produced as size of filter is small.
    
- Prewill:
    less noisy image produced as size of filter is large and 8 filters are used so.
    but still noisy as orginal image has noise.
- Sobel:
    less noisy image produced as size of filter is large and 2 filters are used so.
    but still noisy as orginal image has noise.
- Canny: Best
    non noise image produced, edges are well connected 
    but still noise can be removed with higher sigma for better edge detection
    

```python
# 1. Built-in Sobel, Prewill, Roberts, Canny

img = rgb2gray(io.imread('circuit.tif'))
edge_roberts = roberts(img)
edge_sobel_h = sobel_h(img)
edge_sobel_v = sobel_v(img)
edge_sobel = sobel(img)
edge_prewitt = prewitt(img)
edge_canny = canny(img)
#show_images([edge_roberts,edge_prewitt,edge_sobel,edge_canny],["Roberts","Prewill","Sobel","Canny"])

show_images([img,edge_roberts,edge_prewitt],["Orginal","Roberts","Prewill"])
show_images([img,edge_sobel_h,edge_sobel_v],["Orginal","Sobel_h","Sobel_v"])
show_images([img,edge_sobel,edge_canny],["Orginal","Sobel","Canny"])
```

### Canny low_threshold & high_threshold
- low_threshold :
    Lower bound for hysteresis thresholding (linking edges).
    low_threshold is set to 10% of max by default.
- high_threshold : 
    Upper bound for hysteresis thresholding (linking edges).
    high_threshold is set to 20% of max by default.

```python
edge_canny_thrs= canny(img,low_threshold=30,high_threshold=60)
edge_canny_sigma_thrs= canny(img,sigma = 2,low_threshold=30,high_threshold=60)
show_images([img,edge_canny_thrs,edge_canny_sigma_thrs],["original","canny","canny sigma = 2"])
```

```python
# 2. Custom Sobel (V-H-Both) using conv2d

hx=np.array([ 
    [-1,-2,-1],
    [0,0,0],
    [1,2,1] 
])
hy=np.array([ 
    [-1,0,1],
    [-2,0,2],
    [-1,0,1] 
])
sobel_x = convolve2d(img,hx,boundary='symm')
sobel_y = convolve2d(img,hy,boundary='symm')
sobel = np.sqrt( (sobel_x)**2  +  (sobel_y)**2 )
print("Before Applying thresholding")
show_images([img,sobel_x,sobel_y,sobel],["original","sobel_x","sobel_y","sobel"])
sobel[sobel <= 85] = 0
print("after Applying thresholding")
show_images([img,sobel_x,sobel_y,sobel],["original","sobel_x","sobel_y","sobel"])
```

```python
def sobelEdgeDetection(img):
    hx = np.array([ 
        [-1,-2,-1],
        [0,0,0],
        [1,2,1] 
    ])
    hy = np.array([ 
        [-1,0,1],
        [-2,0,2],
        [-1,0,1] 
    ])
    sobel_x = np.zeros((img.shape[0],img.shape[1]))
    sobel_y = np.zeros((img.shape[0],img.shape[1]))
    sobel = np.zeros((img.shape[0],img.shape[1]))
    for i in range(1,img.shape[0]-1):
        for j in range(1,img.shape[1]-1):
            sobel_x[i,j] = np.sum(img[i-1:i+2,j-1:j+2] * hx )
            sobel_y[i,j] = np.sum(img[i-1:i+2,j-1:j+2] * hy )
            sobel[i,j] =np.sqrt( (sobel_x[i,j])**2  +  (sobel_y[i,j])**2 )
    sobel[sobel <= 25] = 0
    show_images([img,sobel_x,sobel_y,sobel],["original","sobel_x","sobel_y","sobel"])
    return sobel,sobel_x,sobel_y
```

```python
sobel,sobel_x,sobel_y = sobelEdgeDetection(img)
```

```python
# 3. Custom LoG
def LOG(img,thres= 20):
    h4 = np.array([ 
        [0,1,0],
        [1,-4,1],
        [0,1,0] 
    ])
    h8 = np.array([ 
        [-1,-1,-1],
        [-1,8,-1],
        [-1,-1,-1] 
    ])
    newImg = gaussian(img, sigma=1,mode='nearest', cval=0, preserve_range=False, truncate=4.0)
    laplace_4 = np.zeros((img.shape[0],img.shape[1]))
    laplace_8 = np.zeros((img.shape[0],img.shape[1]))
#     for i in range(1,img.shape[0]-1):
#         for j in range(1,img.shape[1]-1):
#             laplace_4[i,j] = np.sum(newImg[i-1:i+2,j-1:j+2] * h4 )
#             laplace_8[i,j] = np.sum(newImg[i-1:i+2,j-1:j+2] * h8 )
    laplace_4 = convolve2d(img,h4,boundary='symm')
    laplace_8 = convolve2d(img,h8,boundary='symm')
    laplace_4[laplace_4 < thres] = 0
    laplace_8[laplace_8 < thres] = 0
    show_images([img,newImg,laplace_4,laplace_8],["original","filtered","laplace 4","laplace 8"])
    return laplace_4,laplace_8
```

```python
laplace_4,laplace_8 = LOG(img,0)
```
