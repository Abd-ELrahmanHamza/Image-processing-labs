```python
img = rgb2gray(io.imread('pout.tif'))
noiseImg = random_noise(img, mode='s&p', seed=None, clip=True,amount = 0.05)
m,n = noiseImg.shape
```

```python
def medianFilter(img,windowHeight,windowWidth):
    m,n = img.shape
    newImg = np.zeros((m,n))
    width = math.floor(windowHeight/2)
    height = math.floor(windowWidth/2)
    for x in range(height,m-height):
        for y in range(width,n-width):
            med = np.median(img[x-height:x+height+1,y-width:y+width+1],axis = None)
            #medianarr = np.sort(img[x-height:x+height+1,y-width:y+width+1])
            #print(medianarr.shape)
            #newImg[x][y] = medianarr[int(windowHeight/2)][int(windowWidth/2)] 
            newImg[x][y] = med 

    return newImg
        
```
